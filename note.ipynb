## Computation: Application to fatigue in composite materials

### Motivation

AI is already widely used in a variaty of fields and applications. Some of which deals with risks and the evaluation of them. These models are black boxes and can show unexpected behaviour, in cases it actually does matter to have a robust model. The point of the paper is to give a method that attempts to quantify these incertainty in the output of the model. Baysen Neural Network gives exactly that but current training methods are quite rigid as they use a parametric system to describe the distribution of the ouput. Although it is good, they did developp a new algorithm that uses a non parametric system for the distribution, which would allow for a model that is more fitted(in the likehood sence) to the reality.

# BNN

Here instead of having point values for the parameters $w, b$, we instead compute a distribution
$ \theta = \{w, b\} \in \Theta \subseteq \mathbb{R}^d$ given a model class $\mathcal{M}$ which here is the network architecture of the neural network

$p (\theta| D, \mathcal{M}) = \frac{p(D| \theta, \mathcal{M}) p(\theta|\mathcal{M})}{p(D|\mathcal{M})}$

* $p(\theta| D, \mathcal{M})$ : posterior distribution of theta <span style="color:red">hard to get</span>
* $p(\theta|\mathcal{M})$ : prior distribution about $\theta$
* $p(D| \theta, \mathcal{M})$ : likelyhood function
* $p(D|\mathcal{M})$ : evidence
