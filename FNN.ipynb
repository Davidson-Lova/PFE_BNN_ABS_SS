$\hat{y}_k = f(x, w, b) = g\left( \sum_{j=1}^m  w_{jk}^{(2)} h(\sum_{i = 1}^n w_{ij}^{(1)} x_i + b_j^{(1)})+ b_k^{(2)}\right)$ 

where
* $x \in \mathcal{X} \subseteq \mathbb{R}^n$ is the input vector

* $\hat{y} \in \mathcal{O} \subseteq \mathbb{R}^l$ is the ouput vector 

* $W^{(1)} = \left(w_{ij}^{(1)}\right)_{1 \leq i \leq n, 1 \leq j \leq m}$ Weight matrix between the input layer and the hidden layer

* $W^{(2)} = \left(w_{jk}^{(2)}\right)_{1 \leq j \leq m, 1 \leq j \leq l}$ weight matrix between the hidden layer and the outpul layer

* $b^{(1)} = \left(b_j^{(1)}\right)_{1 \leq j \leq m}$ biais between the input layer and the hidden layer

* $b^{(2)} = \left(b_k^{(2)}\right)_{1 \leq k \leq l}$ biais matrix between the hidden layer and the outpul layer
* $f,g$ activation functions
